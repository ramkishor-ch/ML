```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#scaling (optional) >> brings all the feature on the same scale
```

```python
#why scaling? --- To reduce the computation time(faster optimisation) and better interpretation
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# Very Important for Interview Questions
# ############# STARTS ##############
```

```python
#standardisation >> zscore = x-mu/sigma
mean = np.mean(df['total_bill'])
```

```python
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
# scaling is optional because scaling doesnt change the distribution of data
# scaling is the best practice to do

#before scaling
sns.distplot(df['total_bill'])
```

```python
#after scaling
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
# Doing Fit One column at a time
scaler.fit(df[['total_bill']]) #fit will calculate mu and sigma
```

```python
# Doing transform One column at a time
scaler.transform(df[['total_bill']]) #apply the params to a formulae(computation)
```

```python
df
```

```python
# Doing Fit and transform 2 columns at a time
scaler.fit_transform(df[['total_bill', 'tip']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill', 'tip']]), columns = ['total_bill', 'tip'])
```

```python
# new data comes into the picture
scaler.transform([[13, 1]])

# New data is considered as test data
```
